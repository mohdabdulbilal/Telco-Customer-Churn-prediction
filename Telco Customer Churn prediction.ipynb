```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import pickle

from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv(r"D:\Data science\projects\Telco Customer Churn prediction\Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['customerID'])
```

```python
df['TotalCharges'] = df['TotalCharges'].map({' ':0})
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df['NoInternetService'] = np.where(df['InternetService'] == 'No', 'Yes', 'No')
```

```python
internet_cols = [
    'StreamingMovies', 'StreamingTV', 'TechSupport',
    'DeviceProtection', 'OnlineBackup', 'OnlineSecurity'
]

df[internet_cols] = df[internet_cols].replace('No internet service', 'No')
```

```python
df['NoPhoneService'] = np.where(df['PhoneService'] == 'No', 'Yes', 'No')

df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
```

```python
if 'PhoneService' in df.columns:
    del df['PhoneService']
```

```python
df.columns
```

```python
df['gender'] = df['gender'].replace(['Male','Female'],[1,0])

cols = ['Partner','Dependents','MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies','PaperlessBilling','NoInternetService',
       'NoPhoneService','Churn']

for col in cols:
    df[col] = df[col].replace(['Yes','No'],[1,0])
```

```python
obj_cols = ['InternetService','Contract','PaymentMethod']

le = LabelEncoder()
for col in obj_cols:
    df[col] = df[col].astype(str) 
    df[col] = le.fit_transform(df[col])
```

```python
num_cols = ['tenure','MonthlyCharges','TotalCharges']
cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
       'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
       'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges',
       'TotalCharges', 'NoInternetService', 'NoPhoneService']
scaler = StandardScaler()
df[cols] = scaler.fit_transform(df[cols])
```

```python
x = df.drop(columns= 'Churn')
y = df['Churn']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', max_iter=1000, solver='lbfgs')
model.fit(x_train, y_train)
```

```python
from imblearn.over_sampling import RandomOverSampler

os = RandomOverSampler()
x_train_res, y_train_res = os.fit_resample(x_train, y_train)

model.fit(x_train_res, y_train_res)
```

```python
pred = model.predict(x_test)

cm = classification_report(y_test,pred)
print(cm)
```

```python
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model saved as model.pkl")
```
